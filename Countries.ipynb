```python
modcountries.head()
```

```python
modcountries[modcountries.isna().any(axis=1)]["Country"]
```

```python
modcountries=modcountries.dropna()
```

```python
y = modcountries["InfantMortality"]
x = modcountries["Deathrate"]

slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept


plt.scatter(x, y, label='Data points', color = "purple")
plt.plot(x, line, color='red', label='Line of best fit')

plt.title('Scatter Plot with Line of Best Fit')
plt.xlabel('Deathrate')
plt.ylabel('InfantMortality')
plt.legend()
plt.show()

equation = f'y = {slope:.2f}x + {intercept:.2f}'

print("Equation of the best-fit line:", equation)
```
